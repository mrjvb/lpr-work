```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('00011.jpg')
h,w = img.shape[:2]
print(w,h)
```

```python
data = open('00011.txt','r').read().split(',')
```

```python
data
```

```python
int(data[0])
```

```python
for i in range(len(data)):
    if(i==0):
        continue
    elif(i<=int(data[0])):
        data[i] = float(data[i])*w
    elif(i<=int(data[0])*2):
        data[i] = float(data[i])*h
```

```python
plt.imshow(img)
```

```python
data
```

```python
pts = np.array( [[data[1],data[5]],
                  [data[2],data[6]],
                  [data[3],data[7]],
                  [data[4],data[8]]], np.int32)
```

```python
pts
```

```python
pts = pts.reshape((-1,1,2))
```

```python
img2 = cv2.polylines(img,[pts],True,(0,255,0))
```

```python
cv2.imwrite('out.png',img2)
```
